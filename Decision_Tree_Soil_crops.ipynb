```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
import pandas as pd
```

```python
sat_train=pd.read_csv("sat_train.csv")
```

```python
sat_train.head()
```

```python
sat_test=pd.read_csv("sat_test.csv")
```

```python
sat_test.head()
```

```python
train_inputs=sat_train.iloc[:,:-1]
train_output=sat_train.iloc[:,-1]
```

```python
test_inputs=sat_test.iloc[:,:-1]
test_output=sat_test.iloc[:,-1]
```

```python
train_output
```

```python
clf=DecisionTreeClassifier(random_state=365,ccp_alpha=0.02)
```

```python
clf=clf.fit(train_inputs,train_output)
pred_test=clf.predict(test_inputs)
```

```python
plot_tree(clf)
```

```python
plot_tree()
```

```python
plt.figure(figsize=(16,5))
plot_tree(clf, filled=True,class_names=["Red Soil","Cotton crop","grey soil","damp gray soil","soil with vegetation stubble",
                                        "mixture class (all types present)","very damp grey soil"])
plt.show()
```

```python
print(classification_report(test_output,pred_test))
```
